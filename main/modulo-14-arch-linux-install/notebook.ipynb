## 1. Despliegue en Entorno Local
En este escenario, tu propia máquina actúa como servidor de archivos para la Máquina Virtual. Es ideal para entornos sin internet o para pruebas rápidas de scripts en desarrollo.

### 🐍 Automatización del Servidor Python
Usaremos un servidor HTTP minimalista que expone los scripts de instalación.

<VSCode.Cell language="python">
import http.server
import socketserver
import socket

PORT = 8080

def get_local_ip():
    s = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
    try:
        s.connect(('10.255.255.255', 1))
        IP = s.getsockname()[0]
    except Exception:
        IP = '127.0.0.1'
    finally:
        s.close()
    return IP

print(f"Lanzando servidor en: http://{get_local_ip()}:{PORT}")

## 2. Configuración y Automatización de Servidor PXE (Nivel Dios)
El Preboot Execution Environment (PXE) permite arrancar un ordenador desde la red. Eliminamos la necesidad de ISOs físicas o pendrives.

### ⚙️ El script `boot.ipxe`
Este archivo es el que "engaña" a la BIOS para que descargue el kernel de Arch desde internet.

<VSCode.Cell language="markdown">
### Flujo de Trabajo iPXE:
1. El equipo arranca y pide una IP (DHCP).
2. El servidor DHCP le entrega la URL del archivo `boot.ipxe`.
3. El equipo lee `boot.ipxe`, descarga `vmlinuz` (Kernel) y `initrd` (Sistema de archivos base) a la RAM.
4. El kernel arranca con el parámetro `script=...`, disparando la autoinstalación.
</VSCode.Cell>

## 3. Infraestructura basada en GitHub (Cloud)
Es el método más escalable. Los scripts viven en un repositorio público y son accesibles mediante URLs directas (Raw).

### 🚀 Comando "Zero-Touch" Universal
```bash
curl -L https://tinyurl.com/arch-weryyy-raw | bash
```

### 📘 Nota Técnica: GitHub Raw vs Web
Cuando automatizamos con `curl`, siempre debemos apuntar a `raw.githubusercontent.com`. Si apuntamos a `github.com/user/repo/blob/...`, descargaremos el HTML de la página (con botones y menús) en lugar del código limpio, lo que romperá la instalación.
